###Distribución LogNormal
Contexto: En una tienda departamental de electrónica, los tiempos de
espera de los clientes para recibir atención en el mostrador de servicio
técnico tienen una distribución lognormal. Se sabe que el tiempo
promedio de espera es de 10 minutos y una desviación estándar de 2
minutos, lo que genera una cola considerable en ciertos momentos del
día.
Objetivo: Generar tiempos de espera para 50 clientes en el mostrador de
servicio y analizar la duración de la fila de espera.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución LogNormal
media = 10  # Tiempo promedio de espera (minutos)
desviacion_estandar = 2  # Desviación estándar (minutos)
clientes = 50  # Número de clientes

# Conversión de parámetros a escala lognormal
mu = np.log(media**2 / np.sqrt(desviacion_estandar**2 + media**2))
sigma = np.sqrt(np.log(1 + (desviacion_estandar**2 / media**2)))

# Generar tiempos de espera
np.random.seed(42)  # Para reproducibilidad
tiempos_espera = np.random.lognormal(mean=mu, sigma=sigma, size=clientes)

# Análisis y resultados
duracion_total_espera = np.sum(tiempos_espera)
espera_promedio = np.mean(tiempos_espera)
print(f"Tiempos de espera generados (en minutos): {tiempos_espera[:10]}")  # Muestra los primeros 10 tiempos
print(f"Duración total de espera (minutos): {duracion_total_espera:.2f}")
print(f"Tiempo promedio de espera por cliente (minutos): {espera_promedio:.2f}")

# Visualización
plt.hist(tiempos_espera, bins=15, color='lightcoral', edgecolor='black')
plt.title("Distribución de tiempos de espera")
plt.xlabel("Tiempo de espera (minutos)")
plt.ylabel("Frecuencia")
plt.show()
```

#CONCLUSION
permite observar una mayoria de clientes con tiempos de espera cercanos a la media, mientras que algunos tendran tiempos significativamente mayores
