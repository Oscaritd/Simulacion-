#SIMULACION
##EJERCICIO 2
Contexto: Una estación de carga rápida para vehículos eléctricos quiere simular el tiempo entre la llegada de los vehículos para prever la cantidad de puntos de carga necesarios durante el día. La llegada de los vehículos sigue una distribución exponencial con una tasa de 1 llegada cada 15 minutos, en promedio. Objetivo: Generar tiempos entre llegadas de vehículos eléctricos durante un día de operación (12 horas).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
tasa_llegadas = 1 / 15  # 1 llegada cada 15 minutos (en llegadas por minuto)
horas_operacion = 12
minutos_operacion = horas_operacion * 60  # Convertimos a minutos

# Generar tiempos entre llegadas
np.random.seed(42)  # Para reproducibilidad
tiempos_entre_llegadas = np.random.exponential(scale=1/tasa_llegadas, size=minutos_operacion)

# Generar tiempos acumulados hasta la última llegada
tiempos_llegada = np.cumsum(tiempos_entre_llegadas)
tiempos_llegada = tiempos_llegada[tiempos_llegada <= minutos_operacion]  # Limitamos a 12 horas

# Resultados
print(f"Cantidad de vehículos simulados: {len(tiempos_llegada)}")
print("Tiempos entre llegadas (en minutos):")
print(tiempos_entre_llegadas[:10])  # Muestra los primeros 10 tiempos generados

# Visualización
plt.hist(tiempos_entre_llegadas, bins=30, color='skyblue', edgecolor='black')
plt.title("Distribución de tiempos entre llegadas")
plt.xlabel("Tiempo entre llegadas (minutos)")
plt.ylabel("Frecuencia")
plt.show()
```

#CONCLUCION
Este análisis permite a los encargados de la estación anticiparse y tomar decisiones informadas sobre la cantidad de puntos de carga necesarios para garantizar un servicio eficiente.
